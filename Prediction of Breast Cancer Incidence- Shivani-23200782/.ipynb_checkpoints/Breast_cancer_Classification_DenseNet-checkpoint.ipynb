# Creation of the Densenet201 Model Architecture

```python
# Defining all the necessary training and testing variables with the train and test dataset.
benign_train = np.array(Dataset_loader('E:\\Summer Semester\\Projects in Math Modeling\\Breast_cancer_proj\\BreaKHis 400X\\train\\benign',224))
malign_train = np.array(Dataset_loader('E:\\Summer Semester\\Projects in Math Modeling\\Breast_cancer_proj\\BreaKHis 400X\\train\\malignant',224))
benign_test = np.array(Dataset_loader('E:\\Summer Semester\\Projects in Math Modeling\\Breast_cancer_proj\\BreaKHis 400X\\test\\benign',224))
malign_test = np.array(Dataset_loader('E:\\Summer Semester\\Projects in Math Modeling\\Breast_cancer_proj\\BreaKHis 400X\\test\\malignant',224))
benign_train_label = np.zeros(len(benign_train))
malign_train_label = np.ones(len(malign_train))
benign_test_label = np.zeros(len(benign_test))
malign_test_label = np.ones(len(malign_test))

# Preparing training and testing datasets
X_train = np.concatenate((benign_train, malign_train), axis=0)
Y_train = np.concatenate((benign_train_label, malign_train_label), axis=0)
X_test = np.concatenate((benign_test, malign_test), axis=0)
Y_test = np.concatenate((benign_test_label, malign_test_label), axis=0)

# Shuffling the data
s = np.arange(X_train.shape[0])
np.random.shuffle(s)
X_train = X_train[s]
Y_train = Y_train[s]

s = np.arange(X_test.shape[0])
np.random.shuffle(s)
X_test = X_test[s]
Y_test = Y_test[s]

# Performing One-hot encoding on the labels
Y_train = to_categorical(Y_train, num_classes=2)
Y_test = to_categorical(Y_test, num_classes=2)

# Splitting the training data into training and validation sets
x_train, x_val, y_train, y_val = train_test_split(
    X_train, Y_train,
    test_size=0.2,
    random_state=11
)

# Plotting some training images
w, h = 60, 40
fig = plt.figure(figsize=(15, 15))
columns = 4
rows = 3

for i in range(1, columns * rows + 1):
    ax = fig.add_subplot(rows, columns, i)
    if np.argmax(y_train[i]) == 0:
        ax.title.set_text('Benign')
    else:
        ax.title.set_text('Malignant')
    plt.imshow(x_train[i], interpolation='nearest')

# Using ImageDataGenerator for data augmentation to handle class imbalancing
train_generator = ImageDataGenerator(
    zoom_range=2,  
    rotation_range=90,
    horizontal_flip=True,  
    vertical_flip=True,  
)

# Building and compiling the model using Functional API
def build_functional_model(backbone, learning_rate=1e-4):
    inputs = Input(shape=(224, 224, 3))
    x = backbone(inputs, training=False)
    x = GlobalAveragePooling2D()(x)
    x = Dropout(0.5)(x)
    x = BatchNormalization()(x)
    outputs = Dense(2, activation='softmax')(x)
    model = Model(inputs, outputs)
    model.compile(
        loss='binary_crossentropy',
        optimizer=Adam(learning_rate=learning_rate),
        metrics=['accuracy']
    )
    return model

# Clearing any previous session and performing Garbage collection
K.clear_session()
gc.collect()

# Loading the pre-trained DenseNet201 model without the top classification layer
resnet = DenseNet201(
    weights='imagenet',
    include_top=False,
    input_shape=(224,224,3)
)

# Building the model with the functional API
model = build_functional_model(resnet, learning_rate=1e-4)
model.summary()
```

# Model Training

```python
# Creating a file path for saving the keras model
filepath = 'E:\\Summer Semester\\Projects in Math Modeling\\Breast_cancer_proj\\BreaKHis 400X\\best_model.keras'

# Setting up the ModelCheckpoint callback with the correct filepath
checkpoint = ModelCheckpoint(filepath, monitor='val_accuracy', verbose=1, save_best_only=True, mode='max')
BATCH_SIZE=8
# Training the model
history = model.fit(
    train_generator.flow(x_train, y_train, batch_size=BATCH_SIZE),
    steps_per_epoch=x_train.shape[0] // BATCH_SIZE,
    epochs=20,
    validation_data=(x_val, y_val),
    callbacks=[checkpoint]
)

# Loading the best model weights saved during training
model.load_weights(filepath)

# Performing predictions on validation data
Y_val_pred = model.predict(x_val)

# Calculating the accuracy score at each epoch
accuracy = accuracy_score(np.argmax(y_val, axis=1), np.argmax(Y_val_pred, axis=1))
print(f"Validation accuracy: {accuracy}")
```

```python
# Custom function to plot the confusion matrix
def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion matrix', cmap=plt.cm.Blues):
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=55)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
   # plt.savefig('confusion_matrix.png')
    plt.show()

cm = confusion_matrix(np.argmax(Y_test, axis=1), np.argmax(Y_pred_tta, axis=1))
cm_plot_label = ['benign', 'malignant']
plot_confusion_matrix(cm, cm_plot_label, title='Confusion Matrix for Breast Cancer Cell Classification')
```

# Conclusion


The DenseNet201 model was utilized to classify breast cancer images as either benign or malignant. The model achieved a high performance across several metrics, including an accuracy of 0.96, precision of 0.91, recall of 0.94, and an F1 score of 0.97. The ROC curve and the corresponding AUC of 0.903, as shown, further affirm the model's effectiveness in distinguishing between the two classes. The confusion matrix indicates that out of 544 predictions, 505 were correct, with 146 true negatives, 359 true positives, 30 false positives, and 9 false negatives.

However, it is important to note that the training accuracy was recorded at 1.0, which implies overfitting has occured. This is due to class imbalance within the dataset, where the number of benign samples are very less compared to the Malignant samples.

# Challenge: Class Imbalance

The observed overfitting is a direct consequence of the class imbalance in the dataset, where malignant class significantly outweighs the benign class. This imbalance causes the model to learn the dominant class better, leading to overfitting and reduced generalization to new data.

# Solution: Data Augmentation

To mitigate the effects of class imbalance, data augmentation techniques are employed. These techniques include the creation of synthetic samples of the benign class through transformations such as rotations, flips, and shifts, effectively increasing the diversity and volume of the training data. By doing so, the model will be better equipped to generalize and avoid overfitting, ultimately improving its performance on unseen data.
